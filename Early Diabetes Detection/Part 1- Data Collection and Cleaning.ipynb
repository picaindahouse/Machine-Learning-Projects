# Part 1: Data Collection and Cleaning

Google: early stage diabetes prediction dataset|

```python
# Import library
import pandas as pd
```

```python
# Read the CSV into a DataFrame  
# Use pd.read_csv("file_name")
df = pd.read_csv("diabetes_data_upload.csv")
df
```

```python
# Replace the values in the DataFrame with data_frame.replace
df = df.replace("No",0)
df = df.replace("Yes",1)

df = df.replace("Positive",1)
df = df.replace("Negative",0)

# Take gender to mean isMale
# Thus if Male- 1, if Female- 0
df = df.replace("Male",1)
df = df.replace("Female",0)
```

```python
df
# As you can see below we have changed all the values to either 0/1 except for Age
```

```python
# Check for missing values
# Use .isnull() to turn all the values into 0 and 1 and use sum on that to find total number of null values
df.isnull().sum()
# Thus as you can see below, no null values
```

```python
# Check the dtypes of the columns
# as long as you get int or float, its good in this scenario
df.dtypes
```

```python
# Renaming columns
# Create a dictionary
replace = {"Gender" : "isMale"}

df = df.rename(columns = replace)
df
# Gender has been renamed to isMale
```

```python
# Getting all the columns names:
df.columns
```

```python
# Now say you want to standardise and get all the column names to be in lower caps:
df.columns.str.lower() #shows us lower case
```

```python
# Thus we can just:
df.columns = df.columns.str.lower() 
df
# Now we have standardised all the column names
```

```python
# Thus, we have collected and cleaned the data
```

```python
# Now we can convert the data frame back into a new csv file:
df.to_csv("diabetes_data_clean.csv", index = None)
```

```python
# We can check to see if the data frame has been converted into csv
pd.read_csv("diabetes_data_clean.csv")
```

Summary
1. Collect the data from UCI Repository
2. Replaced strings to 1s and 0s
3. Replaced changed column name
4. Lowercased everything in columns
5. Exported the clean Data Frame to a new CSV
